```python
import numpy as np
import tensorflow as tf
```

Định nghĩa hàm loss

```python
theta = tf.Variable(-10.0)
epsilon = 0.0001

@tf.function
def Loss(theta):
    return theta**2 - 4*theta + 10 # (theta-2)^2 + 6
```

Lặp<br>
Cập nhật theta

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.01)

while True:
    with tf.GradientTape() as tape:
        loss = Loss(theta)
        print(loss)
    
    # Tính giá trị đạo hàm của hàm loss và gán vào biến grads
    # Ttrong trường hợp tổng quát thì sẽ có nhiều grads (vì cần tính đạo hàm theo từng biến), nên mới đặt tên có thêm chữ "s" cho biến grads
    # Vì thế mới cần một dãy các tham số theta để tính loss, nên mới cần để biến theta vào list
    grads = tape.gradient(loss, [theta])
    print(grads)

    # Cập nhật tham số theta
    opt.apply_gradients(zip(grads, [theta]))

    # Kiểm tra khi nào dừng
    # grads là một dãy số, nhưng ở đây ta chỉ tính đạo hàm cho mỗi một biến theta, nên chỉ cần lấy phần tử đầu tiên (và là duy nhất) ra để so sánh với epsilon
    # Còn trong trường hợp tổng quát thì phải so sánh từng giá trị đạo hàm (nguyên cả grads) với epsilon
    if abs(grads[0].numpy()) < epsilon:
        break
```
