```python
import numpy as np
import tensorflow as tf
```

Định nghĩa hàm loss

```python
theta = tf.Variable(-10.0)
loss = lambda: theta**2 - 4*theta + 10 # (theta-2)^2 + 6
```

Lặp<br>
Cập nhật theta

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.1)

for i in range(1000):
    opt.minimize(loss, var_list=[theta])
```

In kết quả

```python
print("Giá trị theta tối ưu hiện tại: ", theta)
```
