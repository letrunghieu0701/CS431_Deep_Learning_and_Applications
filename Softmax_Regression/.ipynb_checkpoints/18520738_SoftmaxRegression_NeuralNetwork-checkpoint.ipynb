```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
n_class = 4
n_featue = 2

n_sample_red = 50
n_sample_green = 70
n_sample_blue = 80
n_sample_magenta = 50
n_sample = n_sample_red + n_sample_green + n_sample_blue + n_sample_magenta

x_center_red = np.array([[-6], [6]])
x_center_green = np.array([[6], [5]])
x_center_blue = np.array([[6], [-4]])
x_center_magenta = np.array([[-7], [-4]])

x_red = x_center_red + np.random.normal(0, n_featue, size=(2, n_sample_red))
x_green = x_center_green + np.random.normal(0, n_featue, size=(2, n_sample_green))
x_blue = x_center_blue + np.random.normal(0, n_featue, size=(2, n_sample_blue))
x_magenta = x_center_magenta + np.random.normal(0, n_featue, size=(2, n_sample_magenta))

ones = np.ones(n_sample)
x = np.hstack((np.hstack((np.hstack((x_red, x_green)), x_blue)), x_magenta))
X = np.vstack((ones, x)).astype('float64')
X = X.T


y_red = np.zeros(n_sample_red) # label 0
y_green = np.ones(n_sample_green) # label 1
y_blue = 1 + np.ones(n_sample_blue) # label 2
y_magenta = 2 + np.ones(n_sample_magenta)
y = np.hstack((np.hstack((np.hstack((y_red, y_green)), y_blue)), y_magenta))
Y = tf.one_hot(y, n_class, axis=0).numpy().astype('float64')

print("X.shape:", X.shape)
print("Y.shape:", Y.shape)
```

```python
# Use the below line in Jupyter Notebook to view the plot inside notebook, it doesn't work in Google Colab
%matplotlib inline

def plot_data():
    plt.scatter(x_red[0][:], x_red[1][:], c='r', marker = 's', edgecolors='black', linewidth=1)
    plt.scatter(x_green[0][:], x_green[1][:], c='g', marker = 'o', edgecolors='black', linewidth=1)
    plt.scatter(x_blue[0][:], x_blue[1][:], c='b', marker = '^', edgecolors='black', linewidth=1)
    plt.scatter(x_magenta[0][:], x_magenta[1][:], c='m', marker = '<', edgecolors='black', linewidth=1)
    plt.show()

def plot_predicted(X, input_layer, hidden_layer_1, hidden_layer_2):
    # Draw arae being classificated
    for i in range(-15, 15):
        for j in range(-15, 15):
            point = np.array([[1], [i], [j]]).astype('float64').T
            point_class = predict_class(point, input_layer, hidden_layer_1, hidden_layer_2)

            if point_class == 0:
                plt.plot(point[0][1], point[0][2], 'rs', alpha=0.2)
            elif point_class == 1:
                plt.plot(point[0][1], point[0][2], 'gs', alpha=0.2)
            elif point_class == 2:
                plt.plot(point[0][1], point[0][2], 'bs', alpha=0.2)
            elif point_class == 3:
                plt.plot(point[0][1], point[0][2], 'ms', alpha=0.2)
                   
    # Draw data being classificated
    classes = predict_class(X, input_layer, hidden_layer_1, hidden_layer_2)
    for i in range(len(classes)):
        if classes[i] == 0:
#             plt.plot(x[0][i], x[1][i], 'rs')
            plt.scatter(x[0][i], x[1][i], c='r', marker = 's', edgecolors='black', linewidth=1)
        elif classes[i] == 1:
#             plt.plot(x[0][i], x[1][i], 'go')
            plt.scatter(x[0][i], x[1][i], c='g', marker = 'o', edgecolors='black', linewidth=1)
        elif classes[i] == 2:
#             plt.plot(x[0][i], x[1][i], 'b^')
            plt.scatter(x[0][i], x[1][i], c='b', marker = '^', edgecolors='black', linewidth=1)
        elif classes[i] == 3:
#             plt.plot(x[0][i], x[1][i], 'm^')
            plt.scatter(x[0][i], x[1][i], c='m', marker = '<', edgecolors='black', linewidth=1)
    plt.show()
    
plot_data()
```

**Set value:** theta<br>
**Define:** loss function<br>
**Note:** Do not define the derivative of the loss function because tensorflow will take the derivative by itself

```python
input_layer = tf.Variable(np.random.randn(n_featue+1, 16).astype('float64'), trainable=True)
hidden_layer_1 = tf.Variable(np.random.randn(16, 32).astype('float64'), trainable=True)
hidden_layer_2 = tf.Variable(np.random.randn(32, n_class).astype('float64'), trainable=True)

def Loss(input_layer, hidden_layer_1, hidden_layer_2):
    Z_1 = tf.matmul(X, input_layer)
    Z_1_sigmoid = tf.sigmoid(Z_1)

    Z_2 = tf.matmul(Z_1_sigmoid, hidden_layer_1)
    Z_2_sigmoid = tf.sigmoid(Z_2)

    Y_hat = tf.matmul(Z_2_sigmoid, hidden_layer_2)
    Y_hat_soft = tf.nn.softmax(Y_hat, axis=1)
    
    log_Y_hat_soft = tf.math.log(Y_hat_soft)
    
    loss_matrix = tf.transpose(Y) * log_Y_hat_soft

    # Sum loss for each data, then sum all the loss
    return - tf.math.reduce_mean(tf.math.reduce_sum(loss_matrix, axis=1), axis=0)

def predict_class(X, input_layer, hidden_layer_1, hidden_layer_2):
    Z_1 = tf.matmul(X, input_layer)
    Z_1_sigmoid = tf.sigmoid(Z_1)

    Z_2 = tf.matmul(Z_1_sigmoid, hidden_layer_1)
    Z_2_sigmoid = tf.sigmoid(Z_2)

    Y_hat = tf.matmul(Z_2_sigmoid, hidden_layer_2)
    Y_hat_soft = tf.nn.softmax(Y_hat, axis=1)
    
    return np.argmax(Y_hat_soft, axis=1)
```

**Loop:**<br>
<span style="margin-left:2em">Update theta</span><br>

```python
# Use the below line in Jupyter Notebook to view the plot with interactive window, it doesn't work in Google Colab
# %matplotlib qt
# Set limits for 2 axises so that the plot keeps drawing only in a certain space
plt.xlim(-15, 15)
plt.ylim(-15, 15)

print("Ground truth data")
plot_data()

opt = tf.keras.optimizers.SGD(learning_rate=0.01)
epsilon = 0.01
count_epoch = 1
epoch = 1000
while True:
    # Calculate loss
    with tf.GradientTape() as tape:
        loss = Loss(input_layer, hidden_layer_1, hidden_layer_2)
        
    # Update theta
    grads = tape.gradient(loss, [input_layer, hidden_layer_1, hidden_layer_2])
    opt.apply_gradients(zip(grads, [input_layer, hidden_layer_1, hidden_layer_2]))
    print("Epoch: %d Loss: %.8f" %(count_epoch, loss.numpy()))

    # Stopping condition
    # Check if all values in loss is less than epsilon
    # It will probably take a long time to satisfy the above condition, so we will have to use epoch
    if (tf.experimental.numpy.all((tf.math.abs(grads[0]) < epsilon)) and 
        tf.experimental.numpy.all((tf.math.abs(grads[1]) < epsilon)) and
        tf.experimental.numpy.all((tf.math.abs(grads[2]) < epsilon))) or count_epoch == epoch:
        break

    count_epoch += 1
```

```python
plt.xlim(-15, 15)
plt.ylim(-15, 15)
print("Classification before running Softmax Regression")
plot_data()
```

```python
plt.xlim(-15, 15)
plt.ylim(-15, 15)
print("Classification after running Softmax Regression")
plot_predicted(X, input_layer, hidden_layer_1, hidden_layer_2)
```
