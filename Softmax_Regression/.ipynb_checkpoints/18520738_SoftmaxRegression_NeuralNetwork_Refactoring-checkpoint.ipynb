```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import tensorflow as tf
```

```python
def create_training_data(n_class, n_samples_centers):
    n_sample = 0
    x = list()
    y = list()
    
    for ith, (num_sample, center_point) in enumerate(n_samples_centers):
        n_sample += num_sample
        
        x_datas_for_class_ith = center_point + np.random.normal(0, 2, size=(2, num_sample))      
        y_labels_for_class_ith = ith + np.zeros(num_sample) # Adding ith to make the label
        
        x.append(x_datas_for_class_ith)
        y.append(y_labels_for_class_ith)
    
    ones = np.ones(n_sample)
    x = np.hstack(x)
    X = np.vstack((ones, x)).astype('float64')
    
    y = np.hstack(y)
    Y = tf.one_hot(y, n_class, axis=0, dtype = tf.float64)
    
    return X, Y, n_sample
```

```python
n_class = 4
n_feature = 2

n_sample_red = 50
n_sample_green = 70
n_sample_blue = 80
n_sample_magenta = 50
n_sample = n_sample_red + n_sample_green + n_sample_blue + n_sample_magenta

x_center_red = np.array([[-6], [6]])
x_center_green = np.array([[6], [5]])
x_center_blue = np.array([[6], [-4]])
x_center_magenta = np.array([[-7], [-4]])

x_red = x_center_red + np.random.normal(0, n_feature, size=(2, n_sample_red))
x_green = x_center_green + np.random.normal(0, n_feature, size=(2, n_sample_green))
x_blue = x_center_blue + np.random.normal(0, n_feature, size=(2, n_sample_blue))
x_magenta = x_center_magenta + np.random.normal(0, n_feature, size=(2, n_sample_magenta))

ones = np.ones(n_sample)
x = np.concatenate((x_red, x_green, x_blue, x_magenta), axis=1) # Stack matrices horizontally
X = np.vstack((ones, x)).astype('float64')

y_red = np.zeros(n_sample_red) # label 0
y_green = np.ones(n_sample_green) # label 1
y_blue = 1 + np.ones(n_sample_blue) # label 2
y_magenta = 2 + np.ones(n_sample_magenta) # label 3
y = np.hstack((np.hstack((np.hstack((y_red, y_green)), y_blue)), y_magenta))
Y = tf.one_hot(y, n_class, axis=0, dtype = tf.float64)

print("X.shape:", X.shape)
print("Y.shape:", Y.shape)
```

```python
w_i_h1 = tf.Variable(np.random.randn(16, n_feature+1).astype('float64'), trainable=True)
w_h1_h2 = tf.Variable(np.random.randn(32, 16).astype('float64'), trainable=True)
w_h2_o = tf.Variable(np.random.randn(n_class, 32).astype('float64'), trainable=True)

def Loss(w_i_h1, w_h1_h2, w_h2_o):
    # Shape: (n_h1, n_sample) = (n_h1, n_feature) @ (n_feature, n_sample)
    hidden_1 = tf.matmul(w_i_h1, X)
    hidden_1_sigmoid = tf.sigmoid(hidden_1)

    # Shape: (n_h2, n_sample) = (n_h2, n_h1) @ (n_h1, n_sample)
    hidden_2 = tf.matmul(w_h1_h2, hidden_1_sigmoid)
    hidden_2_sigmoid = tf.sigmoid(hidden_2)

    # Shape: (n_class, n_sample) = (n_class, n_h2) @ (n_h2, n_sample)
    Y_hat = tf.matmul(w_h2_o, hidden_2_sigmoid)
    Y_hat_soft = tf.nn.softmax(Y_hat, axis=0)

    log_Y_hat_soft = tf.math.log(Y_hat_soft)

    loss_matrix = Y * log_Y_hat_soft

    # Sum loss for each data, then sum all the loss
    return - tf.math.reduce_mean(tf.math.reduce_sum(loss_matrix, axis=0))

def predict_class(X, w_i_h1, w_h1_h2, w_h2_o):
    hidden_1 = tf.matmul(w_i_h1, X)
    hidden_1_sigmoid = tf.sigmoid(hidden_1)

    hidden_2 = tf.matmul(w_h1_h2, hidden_1_sigmoid)
    hidden_2_sigmoid = tf.sigmoid(hidden_2)

    Y_hat = tf.matmul(w_h2_o, hidden_2_sigmoid)
    Y_hat_soft = tf.nn.softmax(Y_hat, axis=0)
    
    return np.argmax(Y_hat_soft, axis=0)
```

**Set value:** theta<br>
**Define:** loss function<br>
**Note:** Do not define the derivative of the loss function because tensorflow will take the derivative by itself

**Loop:**<br>
<span style="margin-left:2em">Update theta</span><br>

```python
# Use the below line in Jupyter Notebook to view the plot with interactive window, it doesn't work in Google Colab
# %matplotlib qt
# Set limits for 2 axises so that the plot keeps drawing only in a certain space
plt.xlim(-15, 15)
plt.ylim(-15, 15)

print("Ground truth data")
plot_data()

opt = tf.keras.optimizers.SGD(learning_rate=0.01)
epsilon = 0.01
count_epoch = 1
epoch = 1000
while True:
    # Calculate loss
    with tf.GradientTape() as tape:
        loss = Loss(w_i_h1, w_h1_h2, w_h2_o)
        
    # Update theta
    grads = tape.gradient(loss, [w_i_h1, w_h1_h2, w_h2_o])
    opt.apply_gradients(zip(grads, [w_i_h1, w_h1_h2, w_h2_o]))
    print("Epoch: %d Loss: %.8f" %(count_epoch, loss.numpy()))

    # Stopping condition
    # Check if all values in loss is less than epsilon
    # It will probably take a long time to satisfy the above condition, so we will have to use epoch
    if (tf.experimental.numpy.all((tf.math.abs(grads[0]) < epsilon)) and 
        tf.experimental.numpy.all((tf.math.abs(grads[1]) < epsilon)) and
        tf.experimental.numpy.all((tf.math.abs(grads[2]) < epsilon))) or count_epoch == epoch:
        break

    count_epoch += 1
```
