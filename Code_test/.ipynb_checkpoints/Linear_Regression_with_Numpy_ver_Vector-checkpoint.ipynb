```python
import numpy as np
import matplotlib.pyplot as plt
```

Phát sinh dữ liệu

```python
x = np.arange(-5, 5, 0.5)
n_sample = len(x)
noise = np.random.normal(0, 1, n_sample)
Y = 5*x - 6 + noise

# plt.plot(x, Y, "ro")
# plt.show
```

Khởi tạo tham số theta0, theta1, alpha và epsilon

```python
alpha = 0.001
epsilon = 0.0001
theta = np.array([[10], [-5]])

# ones = np.ones((1, n_sample))
# print(ones)
# print(x)
# X = np.concatenate((ones, [x]))
# print(X)
ones = np.ones(n_sample)
X = np.vstack((ones, x)) # Stack arrays in sequence vertically
print("Matix X:\n", X)
print("\nArray Y:\n", Y)
```

Lặp<br>
Cập nhật tham số<br>
Điều kiện dừng

```python
while True:
    nabla = (1.0/n_sample) * np.dot(X, np.transpose(np.dot(np.transpose(theta), X) - Y))
    
    theta = theta - alpha*nabla
    
    # Trực quan hóa kết quả: vẽ phương trình đường thẳng
    x_vis = np.array([-5.0, 5.0])
    y_vis = theta[1][0]*x_vis + theta[0][0]
    plt.plot(x_vis, y_vis)
    
    nabla = (1.0/n_sample) * np.dot(X, np.transpose(np.dot(np.transpose(theta), X) - Y))
    if (nabla[0][0] < epsilon) and (nabla[1][0] < epsilon):
        break

plt.show()
```

In kết quả

```python
print(f"Giá trị tối ưu của theta_1: {theta[0][0]}")
print(f"Giá trị tối ưu của theta_0: {theta[1][0]}")
```
