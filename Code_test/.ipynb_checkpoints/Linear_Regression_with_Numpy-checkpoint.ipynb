```python
import numpy as np
import matplotlib.pyplot as plt
```

Phát sinh dữ liệu

```python
x = np.arange(-5, 5, 0.5)
n_sample = len(x)
noise = np.random.normal(0, 1, n_sample)
y = 5*x - 6 + noise

# plt.plot(x, y, "ro")
# plt.show
```

Khởi tạo tham số theta0, theta1, alpha và epsilon

```python
theta_0 = 10
theta_1 = -4
alpha = 0.001
epsilon = 0.0001
```

Lặp<br>
Cập nhật tham số<br>
Điều kiện dừng

```python
while True:
    DaoHamTheoTheta_0 = np.mean(theta_1*x + theta_0 - y)
    DaoHamTheoTheta_1 = np.mean((theta_1*x + theta_0 - y)*x)
    theta_0 = theta_0 - alpha*DaoHamTheoTheta_0
    theta_1 = theta_1 - alpha*DaoHamTheoTheta_1
    
    # Trực quan hóa kết quả: vẽ phương trình đường thẳng
    x_vis = np.array([-5.0, 5.0])
    y_vis = theta_1*x_vis + theta_0
    plt.plot(x_vis, y_vis)
    
    DaoHamTheoTheta_0 = np.mean(theta_1*x + theta_0 - y)
    DaoHamTheoTheta_1 = np.mean((theta_1*x + theta_0 - y)*x)   
    if (DaoHamTheoTheta_0 < epsilon) and (DaoHamTheoTheta_1 < epsilon):
        break

plt.show()
```

In kết quả

```python
print(f"Giá trị tối ưu của theta_1: {theta_1}")
print(f"Giá trị tối ưu của theta_0: {theta_0}")
```
