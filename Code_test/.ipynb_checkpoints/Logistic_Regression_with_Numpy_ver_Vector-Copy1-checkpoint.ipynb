```python
import numpy as np
import matplotlib.pyplot as plt
```

Phát sinh dữ liệu

```python
n_sample = 15
center_red = np.array([[1], [5]])
center_blue = np.array([[-6], [1]])
x_red = center_red + np.random.normal(0, 1.5, size=(2, n_sample))
x_blue = center_blue + np.random.normal(0, 1.5, size=(2, n_sample))
X = np.hstack((x_red, x_blue))

X = np.vstack((np.ones(n_sample*2), X))
# ones = np.ones(n_sample)
# X = np.vstack((ones, x))
print(X.shape)
print("Ma trận X:\n", X)


Y = np.hstack((np.zeros(n_sample), np.ones(n_sample)))
print("Vector Y:\n", Y)
plt.plot(x_red[0, :], x_red[1, :], "ro")
plt.plot(x_blue[0, :], x_blue[1, :], "bo")
# plt.show
```

Khởi tạo tham số theta0, theta1, alpha và epsilon

```python
alpha = 0.01
epsilon = 0.002
theta = np.array([[1.0], [2.0], [3.0]])
```

Lặp<br>
Cập nhật tham số<br>
Điều kiện dừng

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
nabla = np.dot(X, sigmoid((np.dot(theta.T, X)) - Y).T)
print(nabla)
```

```python
plt.plot(x_red[0, :], x_red[1, :], "ro")
plt.plot(x_blue[0, :], x_blue[1, :], "bo")
while True:
    nabla = np.dot(X, sigmoid((np.dot(theta.T, X)) - Y).T)
    print(nabla)
    theta = theta - alpha*nabla
    
    # Trực quan hóa kết quả: vẽ phương trình đường thẳng
    x_vis = np.array([-8.0, 2.0])
    y_vis = -(theta[0][0]*x_vis + theta[2][0]) / theta[1][0]
    plt.plot(x_vis, y_vis)
    
    nabla = np.dot(X, sigmoid((np.dot(theta.T, X)) - Y).T)
    if abs(nabla[0][0]) < epsilon:
        break

plt.show()
```

```python
plt.plot(x_red[0, :], x_red[1, :], "ro")
plt.plot(x_blue[0, :], x_blue[1, :], "bo")
```
